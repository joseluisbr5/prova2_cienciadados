```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar dataset
df = pd.read_csv("funcionarios_empresa.csv")

# Visualizar primeiras linhas
df.head()

```

```python
df_vendas = df[df['Departamento'] == 'Vendas']
print(df_vendas.shape)
df_vendas.head()

```

```python
# Informações gerais
df_vendas.info()

# Estatísticas descritivas
df_vendas.describe()

# Verificar valores ausentes
df_vendas.isnull().sum()

```

```python
media_salario = df_vendas['Salario'].mean()
mediana_salario = df_vendas['Salario'].median()
desvio_salario = df_vendas['Salario'].std()

print("Média:", media_salario)
print("Mediana:", mediana_salario)
print("Desvio padrão:", desvio_salario)
```

```python
df_vendas['Cargo'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_vendas['Salario'], bins=20, kde=True)
plt.title("Distribuição dos Salários - Departamento de Vendas")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_vendas['Salario'])
plt.title("Boxplot dos Salários - Vendas")
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Cargo", y="Salario", data=df_vendas, estimator=np.mean)
plt.title("Salário Médio por Cargo - Vendas")
plt.xticks(rotation=45)
plt.show()

```
